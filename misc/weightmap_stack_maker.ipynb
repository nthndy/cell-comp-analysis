## weightmap maker

```python
### weightmap maker
for i, name in enumerate(images): 
    # get the current image and make it bin
    image_name = images[i]
    print(image_name)
    image = imread(os.path.normpath(os.path.join(sep_output_path, image_name))) #if previous image sep section not used, replace path with original path
    labelled, n_labels = ndimage.label(image)
    
    weight_mask = np.zeros(image.shape, dtype=np.float32)
    for i in range(1,n_labels+1):
        cell = labelled == i
        not_cell = np.logical_xor(cell, image) #np.logical_and(labelled != i, labelled > 0)
        image_diff = gaussian_filter(cell.astype(np.float32), sigma=5) * gaussian_filter(not_cell.astype(np.float32), sigma=5)
        weight_mask += image_diff

    wmap = w0*weight_mask #* wmap
    
    imsave(os.path.join(weight_output_path, image_name), wmap)
    
    
   
```
