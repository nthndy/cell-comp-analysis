```python
for i in tqdm(new_list):
    GV, pos = i[0:6], i[7:]
    #GV,pos = 'GV0796','Pos8'

    pos_folder = '/media/nathan/Transcend/Nathan/data/h2b/{}/{}/{}_aligned'.format(GV,pos,pos)
    write_folder = '/media/nathan/Transcend/Nathan/data/h2b/{}/{}/stacks'.format(GV,pos)
    if not os.path.exists(write_folder):
        os.mkdir(write_folder)
    for channel in [1,2]:
    # No. Timesteps
        ch = 'GFP' if channel == 1 else 'RFP'
        files = glob.glob(os.path.join(pos_folder,'img_channel00{}*.tif'.format(channel)))
        files.sort(key=lambda f:int(f[-18:-9]))
        images = []
        for tif_ad in files:
            tif = np.array(tiff.imread(tif_ad),dtype=np.uint8)
            tif_w,tif_h = np.shape(tif)[0],np.shape(tif)[1]
            tif = tif[int((tif_w-1200)/2):int(tif_w-(tif_w-1200)/2),int((tif_h-1600)/2):int(tif_h-(tif_h-1600)/2)]
            images.append(tif)
        images = np.stack(images,axis=0)
        with tiff.TiffWriter(os.path.join(write_folder,'{}_{}_{}_8_bit.tif'.format(GV,pos,ch)),imagej=True) as tif_:
            tif_.save(images)
        for file in files:
            os.remove(file)
```
