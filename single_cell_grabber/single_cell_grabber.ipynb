### this script takes a timelapse of segmented cell images and isolates a single mask from them for single cell trajectory analysis



```python
import os
from skimage.io import imread, imsave,imshow
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import ndimage
import re
from natsort import natsorted
```

## Datapaths
##### model is the preexisting single cell trajectories i have manually isolated

```python
input_path = os.path.normpath("C:/Users/Nathan/Documents/data/fucci/labels_v5/")
images = [f for f in os.listdir(input_path) if f.endswith('.tif')]
images = natsorted(images)
#print(images)

model_path = os.path.normpath("C:/Users/Nathan/Documents/data/fucci/singlecell_labels_manual")
model_cell_dirs = [d for d in os.listdir(model_path) if d.startswith('labels') and not d.endswith('unfinished')]

print("model cell trajectories are: ", model_cell_dirs)

output_dir = "singlecell_labels_model_" + version
output_path = os.path.normpath(os.path.join("C:/Users/Nathan/Documents/data/fucci/", output_dir))

try:
    # Create target Directory
        os.mkdir(output_path)
        print("Directory " , output_path ,  " Created ") 
except FileExistsError:
        print("Directory " , output_path ,  " already exists")
        
for m in model_cell_dirs:
    new_cell_dir = os.path.join(output_path, m)
    try:
    # Create target Directory
        os.mkdir(new_cell_dir)
        print("Directory " , new_cell_dir ,  " Created ") 
    except FileExistsError:
        print("Directory " , new_cell_dir ,  " already exists")
```

```python
#get first cell images, take coordinates from them, store coordinates, apply to input
for m, model_dir in enumerate(model_cell_dirs):
    print(model_cell_dirs)
    new_cell_dir = os.path.join(output_path, model_dir)
    #new_cell_dir = os.path.join(output_path, "labels_cell1a")
    print(new_cell_dir)

    model_images = [i for i in os.listdir(os.path.join(model_path, model_cell_dirs[m])) if i.endswith('.modified.tif')]
    print(model_images)
    for n, model_image in enumerate(model_images):
        model_mask = imread(os.path.normpath(os.path.join(model_path, model_cell_dirs[m], model_images[n])))
        #plt.imshow(model_mask)
        coords = ndimage.measurements.center_of_mass(model_mask)
        #now take this coord and open corresponding new_image
        input_image = imread(os.path.normpath(os.path.join(input_path ,images[n])))
        #plt.imshow(input_image)
        labelled, n = ndimage.label(input_image)
        output_array = labelled == labelled[int(coords[0]), int(coords[1])]
        #plt.imshow(output_image)
        #for d in model_cell_dirs:
        output_fn = model_image[:-len('.modified.tif')]
        print(output_fn)
        output_image = output_array * 1
        imsave(os.path.join(new_cell_dir, model_image), output_image)
    
print("donedonedonedonedone")
```
