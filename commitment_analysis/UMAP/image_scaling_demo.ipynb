```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gfp = np.random.random(size=(64, 64))*10.
rfp = np.random.random(size=(64, 64))*10.
```

```python
gfp[:32,:32] = gfp[:32,:32] + 100. 
rfp[32:,32:] = rfp[32:,32:] + 10.
```

```python
plt.imshow(gfp)
```

```python
plt.imshow(rfp)
```

```python
def norm(x):
    return (x-np.mean(x)) / np.max([np.std(x), 1/np.prod(x.shape)])
```

```python
ngfp = norm(gfp)
nrfp = norm(rfp)
```

```python
plt.imshow(ngfp)
plt.colorbar()
```

```python
plt.imshow(nrfp)
plt.colorbar()
```

```python
proj = np.sum(np.stack([ngfp, nrfp], axis=-1), axis=-1)
```

```python
plt.imshow(proj)
```

```python
plt.imshow(np.sum(np.stack([gfp, rfp], axis=-1), axis=-1))
```
