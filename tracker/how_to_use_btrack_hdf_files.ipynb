# how to use `btrack` HDF5 files

```python
import os
import numpy as np
import btrack
```

```python
fn = "/Users/arl/Dropbox/Code/py3/arboretum/segmented.hdf5"
```

## load the file and get the tracks

```python
with btrack.dataio.HDF5FileHandler(fn, 'r', obj_type='obj_type_1') as hdf:
    tracks = hdf.tracks
```

## let's look at their contents

`tracks` is just a list of `Tracklet` objects.

```python
tracks[0]
```

```python
tracks[10].ID
```

```python
filtered_tracks = [track for track in tracks if track.fate.name == "APOPTOSIS"]
```

```python
def euclidean_distance(t0, t1, frame):
    try:
        idx0 = t0.t.index(frame)
        idx1 = t1.t.index(frame)
    except:
        return np.inf
    
    dx = t0.x[idx0] - t1.x[idx1]
    dy = t0.y[idx0] - t1.y[idx1]
    
    return np.sqrt(dx**2 + dy**2)

def find_nearby_cells(target_track, radius=10, delta_t=0):
    frame = target_track.t[-1] + delta_t
    tracks_in_frame = [track for track in tracks if track.in_frame(frame) and track != target_track]
    tracks_in_radius = [track for track in tracks_in_frame if euclidean_distance(target_track, track, frame)<radius]
    return tracks_in_radius
```

```python
tracks_in_window = find_nearby_cells(filtered_tracks[0], radius=150, delta_t=0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,16))
plt.plot(filtered_tracks[0].x, filtered_tracks[0].y, 'r-')
for track in tracks_in_window:
    plt.plot(track.x, track.y, 'b-')
plt.show()
```
